```
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from imutils import paths
from sklearn.metrics import classification_report
```

```
label_name = ["Old", "Middle", "Young"]
```

```
print("Data Loading..")
imagePaths = list(paths.list_images("/content/AGE-ESTIMATION-COMPARISON/dataset/train"))
print(imagePaths)
print(len(imagePaths))
```

```
def load(imagePaths):
      data = []
      labels = []
      for(i, imagePath) in enumerate(imagePaths):
        image = cv2.imread(imagePath)
        label = imagePath.split("/") [-2]
        image = cv2.resize(image,(32,32), interpolation=cv2.INTER_AREA)
        data.append(image)
        labels.append(label)
      return (np.array(data), np.array(labels))
```

```
(data, labels) = load(imagePaths)
print("Data shape",data.shape)
print("Label shape",labels.shape)
```

```
data_flattern = data.reshape(15923,3072)
print(data_flattern.shape)
```

```
trainX = data_flattern
trainY = labels 

trainX,testX,trainY,testY = train_test_split(trainX,trainY,test_size=0.2, random_state=42)
```

```
print("KNN ALgorithm loading")
model = KNeighborsClassifier(n_neighbors=1)
model.fit(trainX, trainY)
```

```
res=model.predict(testX[10].reshape(1,-1))
print(res,testY[10])
```

```
predictions =  model.predict(testX)
```

```
print(classification_report(testY, predictions))
```
